```python
def marked_attendance():
    student = set()
    while True:
       input_name= input("Enter the name of student").strip().lower()
       if input_name=="stop":
           break
       student.add(input_name) 
    print("Students who marked attendance:")
    for name in student:
        print(name) 
    print("Total students marked attendance:", len(student))

marked_attendance()
```
